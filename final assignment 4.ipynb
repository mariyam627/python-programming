```python
# 7. marksheet
eng = int(input("Enter English number: "));
math = int(input("Enter Math number: "));
urdu = int(input("Enter Urdu number: "));
sci = int(input("Enter Science number: "));
isl = int(input("Enter Islamiat number: "));
total = 500;
print("Total Marks: " ,total)
obtmark = eng + math + urdu + sci + isl;
print("Obtained Marks: " ,obtmark);
percentage = obtmark / total * 100;
print("Percentage: " ,percentage);
if (percentage > 100 and percentage > 80):
    print("Grade A+");
elif (percentage > 80 and percentage > 70):
    print("Grade A");
elif (percentage > 70 and percentage > 60):
    print("Grade B");
elif (percentage > 60 and percentage > 50):
    print("Grade C");
elif (percentage > 50 and percentage > 40):
    print("Grade D");
elif (percentage > 40 and percentage > 35):    
    print("Grade E");
else:
    print("Fail");
```

```python
# 10. python program to sum all the numeric items in a list?
numbers = [6,4,3,2]
def sum(numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(sum(numbers));
```
